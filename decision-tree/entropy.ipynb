### Q: How to calculate entropy?

$E(S) = \Sigma_{i=1}^c - p_i{\cdot}log_2{\cdot}p_i$

```python
%matplotlib inline
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
entropy = lambda x: -x * math.log(x, 2) if x > 0 else 0

# Generate a linear space from 0 to 1, with a step of 0.1
repeat = 11
x = np.linspace(0, 1, repeat)
y = [(entropy(x_i) + entropy(1 - x_i)) for x_i in x]
y
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set(xlabel='Probability', ylabel='Entropy', title='Entropy = -p$\cdot$log$_2$p - q$\cdot$log$_2$q')
ax.grid()
```
